### 1. Tỷ lệ tương tác (Engagement Rate): Tính bằng tổng số tương tác (như like, share, comment) chia cho tổng số người theo dõi hoặc tổng số lượt xem

```python
# Nhóm dữ liệu theo cột 'Shop' và tính tổng của các giá trị trong các cột
totals_per_shop = df_tiktok.groupby('Shop')[['diggCount', 'shareCount', 'commentCount', 'playCount', 'collectCount']].sum()

# Hiển thị tổng theo từng shop
print(totals_per_shop)
```

```python
# Tính tổng số tương tác
totals_per_shop['total_interactions'] = totals_per_shop['diggCount'] + totals_per_shop['shareCount'] + totals_per_shop['commentCount'] + totals_per_shop['collectCount']

# Tính tỷ lệ tương tác
totals_per_shop['engagement_rate'] = (totals_per_shop['total_interactions'] / totals_per_shop['playCount']) * 100

# Vẽ biểu đồ
fig, ax1 = plt.subplots(figsize=(12, 8))

# Vẽ biểu đồ cột cho tỷ lệ tương tác
bars = ax1.bar(totals_per_shop.index, totals_per_shop['engagement_rate'], color=['blue', 'green', 'orange'], alpha=0.6, label='Engagement Rate (%)')

# Thêm giá trị lên trên mỗi cột
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2.0, yval + 0.2, f'{yval:.2f}', ha='center', va='bottom')

# Tạo trục phụ để vẽ đường cho số lượng người xem
ax2 = ax1.twinx()
ax2.plot(totals_per_shop.index, totals_per_shop['playCount'], color='red', marker='o', linestyle='-', linewidth=2, label='Play Count')

# Cài đặt các thông số trục và tiêu đề
ax1.set_xlabel('Shop')
ax1.set_ylabel('Engagement Rate (%)', color='black')
ax2.set_ylabel('Play Count', color='red')
ax1.set_title('Engagement Rate (Like, Share, Comment, Collect) and Play Count Comparison')

ax2.set_yscale('log')

ax1.set_ylim(0, totals_per_shop['engagement_rate'].max() + 1)  # Tăng giới hạn trục y chính thêm 4
ax2.set_ylim(0, totals_per_shop['playCount'].max() + 1000)  # Tăng giới hạn trục y phụ thêm giá trị cụ thể (nếu cần)

# Thêm chú thích
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lọc các cột media
media_columns = ['isVideo', 'media/0/__isMedia', 'media/1/__typename', 'media/2/__typename', 'media/3/__typename', 'media/4/__typename', 'media/5/__typename']

# Tạo một cột mới hợp các cột media
def combine_media(row):
    media_values = [row[col] if pd.notna(row[col]) else 'Video' for col in media_columns]
    media_values = [item for item in media_values if item != 'Video']
    return media_values[0] if len(media_values) > 0 else 'Video'

df_facebook['CombinedMedia'] = df_facebook.apply(combine_media, axis=1)
```
